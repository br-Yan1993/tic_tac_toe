```python
from tkinter import *
import random

janela = Tk()

janela.geometry("200x250+540+360")
janela.config(background="cyan")

lista_locais = []
lista_produto_posicao = [2,2,2,2,2,2,2,2,2]

a1 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(a1))
a1.place(x=10,y=60)

a2 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(a2))
a2.place(x=60,y=60)

a3 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(a3))
a3.place(x=110,y=60)

b1 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(b1))
b1.place(x=10,y=110)

b2 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(b2))
b2.place(x=60,y=110)

b3 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(b3))
b3.place(x=110,y=110)

c1 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(c1))
c1.place(x=10,y=160)

c2 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(c2))
c2.place(x=60,y=160)

c3 = Button(text=" ", height=2, width=4, background="white", command=lambda : setar(c3))
c3.place(x=110,y=160)

lista_botoes = [a1,a2,a3,b1,b2,b3,c1,c2,c3]

### Funções ###

def setar(botao):
    botao.configure(text="X", state=DISABLED)
    lista_locais.append(botao)
    print(lista_locais)
    print(len(lista_locais))

    primeira_rodada(botao)

def primeira_rodada(botao):
    if len(lista_locais) == 1 and lista_locais.count(b2) == 1:
        pos_aleatoria = random.choice([a1, a3, c1, c3])
        pos_aleatoria.configure(text="O", state=DISABLED)
        lista_locais.append(botao)
    elif len(lista_locais) == 1 and lista_locais.count(b2) == 0:
        botao = b2
        botao.configure(text="O", state=DISABLED)
        lista_locais.append(botao)
    else:



def analise_jogo():
    for i in range(len(lista_botoes)):
        if lista_botoes[i].cget("text") == "X":
            lista_produto_posicao[i] = 3
        elif lista_botoes[i].cget("text") == "O":
            lista_produto_posicao[i] = 4




    

janela.mainloop()
```
